# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):
    return abs(x) * (x > 0)
def dReLU(x):
    return (1 * (x > 0))

x=plt.linspace(-10,10,15)
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,15)')
```

```python
plt.plot(x, dReLU(x), 'b', label='linspace(-10,10,15)')
```

```python
plt.plot(x,ReLU(x),'r')
plt.text(6, 8, r'$f(x)=max(0,x)$', fontsize=15)
plt.plot(x,dReLU(x),'b')
plt.text(6, 2, 'x < 0 ,x = 0')
plt.text(6, 1.5, 'x >= 0 ,x = 1')
```
