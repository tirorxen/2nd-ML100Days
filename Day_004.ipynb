```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path, 並讀取 app_train
dir_data = './data/'
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

## 作業
將下列部分資料片段 sub_train 使用 One Hot encoding, 並觀察轉換前後的欄位數量 (使用 shape) 與欄位名稱 (使用 head) 變化

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train.head()
```

```python
sub_train = pd.get_dummies(sub_train)
print(sub_train.shape)
sub_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

gle = LabelEncoder()
for i in range(0 , app_train.shape[1]):
    if app_train.iloc[: , i].dtype == 'object':
        try:
            genre_labels = gle.fit_transform(app_train.iloc[: , i])
            temp = pd.DataFrame(genre_labels)
            app_train = pd.concat([app_train , temp] , axis = 1)
            col_names = app_train.columns.values
            col_names[-1] = app_train.columns[i] + 'label_encoding'
            app_train.columns = col_names 
        except TypeError:
            app_train.iloc[: , i] = app_train.iloc[: , i].apply(lambda x : str(x))
            genre_labels = gle.fit_transform(app_train.iloc[: , i])
            temp = pd.DataFrame(genre_labels)
            app_train = pd.concat([app_train , temp] , axis = 1)
            col_names = app_train.columns.values
            col_names[-1] = app_train.columns[i] + 'label_encoding'
            app_train.columns = col_names
            
app_train.head()
```
