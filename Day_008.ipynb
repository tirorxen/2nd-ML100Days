```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 作業
1. 請將 app_train 中的 CNT_CHILDREN 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_CHILDREN_GROUP
    * 0 個小孩
    * 有 1 - 2 個小孩
    * 有 3 - 5 個小孩
    * 有超過 5 個小孩

2. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，列出各組的平均 AMT_INCOME_TOTAL，並繪製 baxplot
3. 請根據 CNT_CHILDREN_GROUP 以及 TARGET，對 AMT_INCOME_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```python
#1
"""
Your code here
"""
print('before' , '\n' , app_train['CNT_CHILDREN'].value_counts())
print('before' , app_train['CNT_CHILDREN'].value_counts().sum())
#app_train.sort_values(by = 'CNT_CHILDREN')
def children_num(dataframe):
    series = dataframe['CNT_CHILDREN']
    if series.iloc[0] == 0 :
        dataframe['CNT_CHILDREN_GROUP'] = '0 children'
    if series.iloc[0] >= 1 and series.iloc[0] <= 2:
        dataframe['CNT_CHILDREN_GROUP'] = '1 ~ 2 children'  
    if series.iloc[0] >= 3 and series.iloc[0] <= 5:
        dataframe['CNT_CHILDREN_GROUP'] = '3 ~ 5 children'
    if series.iloc[0] > 5:
        dataframe['CNT_CHILDREN_GROUP'] = 'over 5 children'
    return dataframe   
temp = app_train.groupby('CNT_CHILDREN').apply(lambda x : children_num(x))
app_train = temp

print('-----------------------------------------------')
print('after' , '\n' , app_train['CNT_CHILDREN_GROUP'].value_counts())
print('after' , app_train['CNT_CHILDREN_GROUP'].value_counts().sum())
```

```python
#2-1
grp = ['CNT_CHILDREN_GROUP','TARGET']

grouped_df = app_train.groupby(grp)['AMT_INCOME_TOTAL']
grouped_df.mean()
```

```python
#2-2
"""
Your code here
"""
plt_column = 'AMT_INCOME_TOTAL'
plt_by = ['CNT_CHILDREN_GROUP' , 'TARGET']

app_train.boxplot(column=plt_column, by = plt_by, showfliers = False, figsize=(12,12))
plt.suptitle('')
plt.show()
```

```python
#3
app_train = app_train.sort_values(by = ['CNT_CHILDREN_GROUP' , 'TARGET'])
def z_transform(dataframe):
    series = dataframe['AMT_INCOME_TOTAL']
    z_trans = (series - series.mean()) / series.std() 
    return z_trans
temp = app_train.groupby(['CNT_CHILDREN_GROUP' , 'TARGET']).apply(lambda x : z_transform(x))

app_train = app_train.reset_index(drop = True)
temp = pd.DataFrame(temp.reset_index(drop = True))
temp.columns = ['AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']
app_train = pd.concat([app_train , temp] , axis = 1)

app_train[['AMT_INCOME_TOTAL' , 'AMT_INCOME_TOTAL_Z_BY_CHILDREN_GRP-TARGET']].head()
```
